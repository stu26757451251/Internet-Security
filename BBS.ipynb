```python
print("請輸入兩個質數，但 mod 4 = 3")
p,q = map(int, input().split())
n = p*q

print("請再輸入另一數，且此數與p,q互質")
s = int(input())
```

```python
#  x_list = xi (for i = 0,1,2,3,4,...,n)
#  output = binary string


for k in range(1,9):
    
    print("k =",k)
    x = []
    output = []
    x.append(s*s%n)
    
    for i in range(1,1000*k+1):
        x.append((x[i-1]*x[i-1])%n)
        output.append(str(x[i]%2))
    
    # uniform distribution
    uni_dis(output,k)
```
